**Método dropna(subset=)**

Exclui do DF registros nulos

```python
dados_sem_vazio.dropna(subset=colunas_drop, inplace=True)
dados_sem_vazio
```

```python
df_sem_nulos = dados_sem_vazio.dropna(subset=colunas_drop).copy()
df_sem_nulos.head()
```

```python
df_sem_nulos.reset_index(drop=True, inplace=True)
df_sem_nulos.head()
```

```python
df_sem_nulos.isna().sum()
```

# LIDANDO COM OUTLIERS

Dados que fogem muito dos valores do DF

imagem_2025-06-27_182731817.png

```python
df_sem_nulos.describe()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=df_sem_nulos['cliente.tempo_servico'])
```

```python
# Primeiro Quartil
Q1 = df_sem_nulos['cliente.tempo_servico'].quantile(0.25)
# Terceiro Quartil
Q3 =  df_sem_nulos['cliente.tempo_servico'].quantile(0.75)
# Interquartil
IQR = Q3 - Q1

Limite_Inferior = Q1 - 1.5 * IQR
Limite_Superior = Q3 + 1.5 * IQR
```

```python
outliers_index = (df_sem_nulos['cliente.tempo_servico'] < Limite_Inferior) | (df_sem_nulos['cliente.tempo_servico'] > Limite_Superior)
```

```python
df_sem_nulos[outliers_index][
    'cliente.tempo_servico'
]
```

**Substituindo valores para os outliers**

```python
df_sem_outliers = df_sem_nulos.copy()
```

```python
# Mesma saida
df_sem_outliers[outliers_index][
    'cliente.tempo_servico'
]
```

```python
df_sem_outliers.loc[outliers_index,'cliente.tempo_servico'] = np.ceil(
    df_sem_outliers.loc[outliers_index,'conta.cobranca.Total'] /
    df_sem_outliers.loc[outliers_index,'conta.cobranca.mensal']
)
```

```python
sns.boxplot(x=df_sem_outliers['cliente.tempo_servico'])
```

```python
df_sem_outliers[outliers_index][
   ['cliente.tempo_servico','conta.cobranca.mensal','conta.cobranca.Total']
]
```

**Removendo valores que continuam sendo outliers**

Ctrl + Shift + L --> seleciona todos os textos iguais

```python
# Primeiro Quartil
Q1 = df_sem_outliers['cliente.tempo_servico'].quantile(0.25)
# Terceiro Quartil
Q3 =  df_sem_outliers['cliente.tempo_servico'].quantile(0.75)
# Interquartil
IQR = Q3 - Q1

Limite_Inferior = Q1 - 1.5 * IQR
Limite_Superior = Q3 + 1.5 * IQR

outliers_index = (df_sem_outliers['cliente.tempo_servico'] < Limite_Inferior) | (df_sem_outliers['cliente.tempo_servico'] > Limite_Superior)
outliers_index
```

```python
df_sem_outliers[outliers_index]
```

**Utilizando '~'**

Funciona para selecionar todas as as linhas, negando as que forem true

```python
df_sem_outliers = df_sem_outliers[~outliers_index]
df_sem_outliers
```

```python
# Amostras Outliers retiradas com sucesso
sns.boxplot(x=df_sem_outliers['cliente.tempo_servico'])
```

```python
df_sem_outliers.reset_index(drop=True,inplace=True)
df_sem_outliers
```

```python
df_sem_id[colunas] = df_sem_id[colunas].replace(mapeamento)
df_sem_id
```
